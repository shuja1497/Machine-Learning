```python
import numpy as np
import pandas as pd
from matplotlib import  pyplot as plt
%matplotlib inline 
import sklearn
from sklearn.decomposition import PCA
```

```python
ds =pd.read_csv('train.csv')
all_data = ds.values[:5000, :]
print all_data.shape
```

```python
X = all_data[:, 1:]
Y = all_data[:, 0]
```

```python
pca = PCA(n_components=X.shape[1])#PCA object and what should be the dim after trans

pca.fit(X)
```

```python
pca.components_.shape # shape of eigen vector matrix
#eiggen values are along diagnols of this matrix
```

```python
pca.get_covariance()
```

```python
#to analize we plot it as an image ..
#non diag will be black
```

```python
plt.figure(0)

plt.imshow(pca.get_covariance(), cmap='gray' , interpolation ='none')#in this matrix not showing a proper result
# therefore try to normalize the data before applying pca ..that is mu =0 and sigma=1
plt.show()
```

```python
#comparing with original covariancce matrix 
```

```python
plt.figure(1)
plt.imshow(np.cov(X.T), cmap='gray' , interpolation ='none')
plt.show()
```

```python
#matrix after transformation
X_ = pca.transform(X) # its X.v
```

```python
c = np.cov(X_.T)
print c[100:110,100:]

c1 = np.cov(X.T)
print c1[100:110,100:]
```

```python
pca.explained_variance_ # contains all eigen values in sorted decreasing order
```

```python
#we will find cumulative fequency ditr
```

```python
#or find the ratio .. cz func already provided by pca
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(0)
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python
f = np.cumsum(pca.explained_variance_ratio_) # f is cumulative sum 
print f.shape
```

```python
# according to cumsum we will decide till how much it should be clipped 
```

```python

plt.figure(0)
plt.ylim(0, 1.01)
plt.plot(f)
plt.show()
```

```python
# we want 95 % to be retained then ::>>>
```

```python
for ix  in range(f.shape[0]):
    if f[ix] >=.95:
        print ix ,f[ix]
        break
```
